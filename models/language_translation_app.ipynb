```python
# Import necessary libraries
from langchain_ollama.llms import OllamaLLM
from langchain_core.prompts import PromptTemplate
import streamlit as st

```

```python
# Step 1: Define the Llama Translator class

class LlamaTranslator:
    def __init__(self, model="llama3.2"):
        self.model = OllamaLLM(model=model)

    def translate(self, text, source_lang, target_lang):
        # Define the prompt template for translation
        template = f"Translate the following text from {source_lang} to {target_lang}: {text}"
        
        # Create a prompt template for Llama 3.2
        prompt = PromptTemplate(
            input_variables=["text", "source_lang", "target_lang"],
            template=template
        )
        
        # Use the model to generate a translation
        translated_text = self.model.predict(text=text, source_lang=source_lang, target_lang=target_lang)
        return translated_text


# Step 2: Define the language codes

LANGUAGE_CODES = {
    "English": "en",
    "Spanish": "es",
    "French": "fr",
    "German": "de",
    "Chinese": "zh",
    "Hindi": "hi",
    "Italian": "it",
    "Portuguese": "pt",
    "Russian": "ru"
}

# Step 3: Build the Streamlit UI for the language translation

def main():
    # Initialize the translator
    translator = LlamaTranslator()

    # Streamlit UI
    st.title("Language Translation App")
    st.markdown("Translate text from one language to another using Llama 3.2")

    # User input for text
    text = st.text_area("Enter text to translate:", "Hello, how are you?")
    
    # Language selection for source and target languages
    source_lang = st.selectbox("Select source language", list(LANGUAGE_CODES.keys()))
    target_lang = st.selectbox("Select target language", list(LANGUAGE_CODES.keys()))
    
    if st.button("Translate"):
        # Translate the text
        translation = translator.translate(text, source_lang, target_lang)
        st.subheader("Translated Text:")
        st.write(translation)

if __name__ == "__main__":
    main()
```
